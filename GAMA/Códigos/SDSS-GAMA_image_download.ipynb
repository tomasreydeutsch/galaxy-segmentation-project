```python
import pandas as pd
import numpy as np
import requests
import urllib
from google.colab import drive

drive.mount('/content/drive')

gama = pd.read_csv('/content/drive/MyDrive/galaxy-segmentation-project/GAMA/GAMA_1st_query.csv')
gama = gama[['cataid', 'ra', 'DEC']][np.logical_and(gama['z']<0.15, gama['mass_stellar_best_fit']>10**9)].reset_index()
```

```python
# response = urllib.request.urlretrieve(img_url, '/content/drive/MyDrive/galaxy-segmentation-project/GAMA/SDSS images/pruebaaa.jpg')
count = 0
for ind in range(len(gama)):
  cataid, ra, dec = gama['cataid'][ind], gama['ra'][ind], gama['DEC'][ind]
  img_url = f'https://skyserver.sdss.org/dr17/SkyServerWS/ImgCutout/getjpeg?ra={ra}&dec={dec}&width=525&height=525&scale=0.099'
  urllib.request.urlretrieve(img_url, f'/content/drive/MyDrive/galaxy-segmentation-project/GAMA/SDSS images/{cataid}.jpg')
  print(count)
  count += 1
```
